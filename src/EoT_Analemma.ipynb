# Equation of time & Analemma
***

This notebook will calculate the equation of time (EoT) for a given location (coordinates) on Earth.
The EoT describes the discrepancy between two kinds of solar time -- the apparent solar time (diurnal mean of the Sun) and the mean solar time (theoretical mean Sun with uniform motion along the celestial equator) [__[Wikipedia: Equation of Time ](https://en.wikipedia.org/wiki/Equation_of_time#Calculation)__]

I have been able to find two different equations for the EoT from two different sources. They each serve a purpose; the first equation is a bit more more precise as it is based on precise astronomical models and uses constants such as the eccentricity and perihelion details, however the second one is a bit more simplistic and can be used for quick computations

$$\text{EoT} = -2e \sin M + y \sin \left(2M+ 2 \lambda_{\text{p}} \right) \quad [\text{Source: Wikipedia}]$$

where _M_ is the mean anomaly of the Earth's orbit, _e_ is Earth's orbital eccentricity, $\lambda$ is the perihelion longitude and _y_ is a factor that accounts for obliquity.

Given most of these are constants, we can arrive at

$$\text{EoT} = -7.659 \sin M + 9.863\sin \left(2M + 3.5932\right) \, \text{minutes.}$$

The other equation is [Source: "Astronomy on the Personal Computer" by Oliver Montenbruck and Thomas Pfleger]

$$\text{EoT} = 7.5 \sin \left( 2B \right) - 9.87\left(4B \right) - 1.914 \sin g + C$$

where

$$B = \frac{360}{365}\left(n-81\right) \left(\text{degrees}\right), \quad \text{where }n \text{ is the day of the year} \\
g \, \left(\text{mean anomaly}\right) = \frac{360}{365.24}\left(n - 81 \right) + \text{offset}$$

An analemma showing the position of the Sun in the sky as seen from a fixed location on Earth at the same mean solar time, as that position varies over the course of a year can be created by calculating the EoT for every day of the year at a given location.
To create an analemma we need to combine the EoT with the solar declination angle, $\delta$ [__[Source: Solarsena.com](https://solarsena.com/solar-declination-angle-calculator/)__]

$$\delta = \arcsin \left[ \sin \left(-23.44^{\circ}\right)\cos \left( \frac{360}{365.24}\left(d + 10\right)+\frac{360}{\pi}\times 0.0167\sin \left(\frac{360}{365.24}\left(d-2\right)\right)\right)\right]$$

The duration of daylight depend on the latitude ($\phi$) of the targeted location via the following [__[Source: Physics.stackexchange.com](https://physics.stackexchange.com/questions/28563/hours-of-light-per-day-based-on-latitude-longitude-formula)__]

$$ H = 2 \arccos \left(-\tan\left(\phi\right)\tan\left(\delta\right)\right) \frac{24 \, \text{h}}{360} $$

The total daylight duration is proportional to 2H as this covers the time from sunrise to sunset, thus

$$\text{Daylight hours} = \frac{2H}{15}$$

since

$$1 \, \text{hour} = \frac{360^{\circ}}{24 \, \text{h}} = 15 \, \text{deg h}^{-1}.$$

### Comparison of the two equations

The following cells will calculate and plot the differences between the two methods and comment upon them afterwards.

```python
# Equation of time
def eot_precise(day_of_year):
    M = 2 * np.pi * (day_of_year - 3) / 365.24  # Note conversion between degrees in markdown text and radius here!!
    return -7.659 * np.sin(M) + 9.863 * np.sin(2 * M + np.radians(3.5932)) # Result in minutes

def eot_simplified(day_of_year):
    B = 2 * np.pi * (day_of_year - 81) / 365     # Note conversion between degrees in markdown text and radius here!!
    g = 2 * np.pi * (day_of_year - 81) / 365.24  # Note conversion between degrees in markdown text and radius here!!
    return 7.5 * np.sin(2 * B) - 9.87 * np.sin(4 * B) - 1.914 * np.sin(g) # Result in minutes

data = []
for day in range(1, days_in_year + 1):
    precise = eot_precise(day)
    simplified = eot_simplified(day)
    data.append({"Day": day, "Precise": precise, "Simplified": simplified})

# Convert to DataFrame
df = pd.DataFrame(data)
df["Difference"] = df["Precise"] - df["Simplified"]

# Plot the results
plt.plot(df["Day"], df["Precise"], label = "Precise EoT")
plt.plot(df["Day"], df["Simplified"], label = "Simplified EoT", linestyle = '--')
plt.plot(df["Day"], df["Difference"], label = "Difference", alpha = 0.85, linestyle = '-.')
plt.axhline(0, color = 'black', linewidth=0.8, linestyle = ':')
plt.xlabel("Day of year")
plt.ylabel("EoT [min]")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
max_deviation = max(df["Difference"])
min_deviation = min(df["Difference"])
avg_deviation = np.mean(df["Difference"])
std = np.std(df["Difference"])

print(f"Maximum deviation: {max_deviation:.2f}")
print(f"Minimum deviation: {min_deviation:.2f}")
print(f"Average deviation: {avg_deviation:.2f}")
print(f"Standard deviation: {std:.2f}")
```

From the results achieved we see clear deviations between the two methods. However, throughout an entire year the deviations cancels out. Thus if you want to look at the EoT throughout an entire year, the method of choice does not matter, but when only looking at a given window period, the chosen method plays an important role.

### Analemma

In the following cells, I am going to calculate the solar declination angle and use that together with the previous EoT to create an analemma.

```python
def calculate_declination(day_of_year):
    return np.arcsin(np.sin(np.radians(-23.44)) * np.cos(2 * np.pi * (day_of_year + 10) / 365.24 + 2 * 0.0167 * np.sin(2 * np.pi * (day_of_year - 2) / 365.24)))

def calculate_daylight(lat, day_of_year):
    decl = calculate_declination(day_of_year)
    lat_rad = np.radians(lat)
    hour_angle = np.arccos(-np.tan(lat_rad) * np.tan(decl))
    daylight_hours = 2 * np.degrees(hour_angle) / 15
    return daylight_hours

def julian_date(year, month, day):
    dt = datetime(year, month, day)
    julian_start = datetime(year - 1, 12, 31, 12)
    return (dt - julian_start).days + 1

def generate_analemma(lat, year):
    days = np.arange(1, 366)
    eot = np.array([eot_precise(d) for d in days])
    declination = np.array([calculate_declination(d) for d in days])
    daylight = np.array([calculate_daylight(lat, d) for d in days])
    julian = np.array([julian_date(year, 1, 1) + d for d in days])
    seasons = [calculate_season(day, year) for day in days]

    return pd.DataFrame({
        "Day": days,
        "JulianDate": julian,
        "EoT": eot,
        "Declination": np.degrees(declination),
        "Daylight": daylight,
        "Season": seasons
    })

def calculate_season(day_of_year, year):
    # Define the approximate dates for equinoxes and solstices
    spring_start = datetime(year, 3, 20).timetuple().tm_yday
    summer_start = datetime(year, 6, 21).timetuple().tm_yday
    fall_start = datetime(year, 9, 22).timetuple().tm_yday
    winter_start = datetime(year, 12, 21).timetuple().tm_yday
    next_winter_start = datetime(year + 1, 3, 20).timetuple().tm_yday + 365

    if spring_start <= day_of_year < summer_start:
        return "Spring"
    elif summer_start <= day_of_year < fall_start:
        return "Summer"
    elif fall_start <= day_of_year < winter_start:
        return "Fall"
    else:
        return "Winter"

def generate_seasons(year):
    days = range(1, 366)  # For non-leap years
    seasons = [calculate_season(day, year) for day in days]
    return pd.DataFrame({"Day of Year": days, "Season": seasons})

TFO_data = generate_analemma(lat = locations["Tenerife"]["latitude"], year = 2024)
MTK_data = generate_analemma(lat = locations["MtKent"]["latitude"], year = 2024) 
```

```python
# Plotting
def plot_analemma(data):
    plt.scatter(data["EoT"], data["Declination"], c=data["Day"], cmap="viridis", s=10)
    plt.colorbar(label="Day of Year")
    plt.xlabel("Equation of Time (minutes)")
    plt.ylabel("Solar Declination (degrees)")
    plt.show()

def plot_day_night(data):
    plt.plot(data["JulianDate"], data["Daylight"], label="Hrs daylight")
    plt.xlabel("Time of year (days)")
    plt.ylabel("Daylight Duration (hours)")
    plt.show()

def plot_analemma_with_seasons(data):
    seasons = data["Season"].unique()
    colors = {"Spring": "green", "Summer": "orange", "Fall": "brown", "Winter": "blue"}
    
    for season in seasons:
        subset = data[data["Season"] == season]
        plt.scatter(subset["EoT"], subset["Declination"], label=season, alpha=0.7, color=colors[season])
    
    plt.xlabel("Equation of Time (minutes)")
    plt.ylabel("Solar Declination (degrees)")
    plt.legend()
    plt.show()

plot_analemma(TFO_data)
plot_analemma_with_seasons(MTK_data)
plot_day_night(TFO_data)
plot_day_night(MTK_data)
```
